```python
import pandas as pd

file_path = "/content/drive/MyDrive/DACON/대구 교통사고 피해 예측AI/open/external_open/countrywide_accident.csv"

df = pd.read_csv(file_path)

df.head(5)
```

```python
df.info()
```

```python
accident_types1 = df['사고유형'].unique()
print(accident_types1)

accident_types2 = df['기상상태'].unique()
print(accident_types2)

accident_types3 = df['사고유형 - 세부분류'].unique()
print(accident_types3)

accident_types4 = df['법규위반'].unique()
print(accident_types4)

accident_types5 = df['피해운전자 상해정도'].unique()
print(accident_types5)

accident_types6 = df['가해운전자 상해정도'].unique()
print(accident_types6)

accident_types7 = df['피해운전자 차종'].unique()
print(accident_types7)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# 기상 상태별로 각 사고 유형의 합계를 계산
df_totals_by_weather = df.groupby('기상상태')[['사망자수', '중상자수', '경상자수', '부상자수']].sum()

# 이 데이터를 재구성하여 시각화에 적합하게 만들기
df_totals_by_weather = df_totals_by_weather.reset_index().melt(id_vars='기상상태', var_name='사고 유형', value_name='사고 수')

# 시각화

plt.figure(figsize=(12,6))
plt.rc('font', family='NanumBarunGothic')
sns.barplot(x='기상상태', y='사고 수', hue='사고 유형', data=df_totals_by_weather)
plt.title('기상상태별 사고 합계')
plt.ylabel('사고 합계')
plt.xlabel('기상상태')
plt.show()

```

```python
df_filtered = df[df['기상상태'] != '맑음']

# 이제 이 필터링된 DataFrame을 사용하여 기상 상태별로 각 사고 유형의 합계를 계산
df_totals_by_weather_filtered = df_filtered.groupby('기상상태')[['사망자수', '중상자수', '경상자수', '부상자수']].sum()

# 데이터 재구성
df_totals_by_weather_filtered = df_totals_by_weather_filtered.reset_index().melt(id_vars='기상상태', var_name='사고 유형', value_name='사고 수')

# 시각화
plt.figure(figsize=(12,6))
sns.barplot(x='기상상태', y='사고 수', hue='사고 유형', data=df_totals_by_weather_filtered)
plt.title('맑음을 제외한 기상상태별 사고 합계')
plt.ylabel('사고 합계')
plt.xlabel('기상상태')
plt.show()
```
